## RGB Channels

```python
plt.imshow(img_array[:,:,0], cmap="Reds", vmin=0, vmax=255)
```

```python
plt.imshow(img_array[:,:,1], cmap="Greens", vmin=0, vmax=255)
```

```python
plt.imshow(img_array[:,:,2], cmap='Blues', vmin=0, vmax=255)
```

```python
img_arr_2 = img_array.copy()
# intensify Red
img_arr_2[:, :, 0] = img_arr_2[:,:,0] * 0
# intensify Green
img_arr_2[:, :, 1] = img_arr_2[:,:,1] * 1
# Blue
img_arr_2[:, :, 2] = img_arr_2[:,:,1] * 0
img_arr_2
```

```python
plt.figure()

#subplot(r,c) provide the no. of rows and columns
f, axarr = plt.subplots(1, 2) 
axarr[0].imshow(img_array)
axarr[1].imshow(img_arr_2)
```

## Crop

```python
plt.imshow(img_array[100:200,100:200,:])
```

## Grayscale

### Solution 1: Average

```python
img_gray = np.average(img_array, axis=-1)
img_gray.shape
```

```python
plt.imshow(img_gray, cmap='gray')
```

### Solution 2: MinMax

```python
img_gray_min_max = np.max(img_array, axis=-1)/2 + np.min(img_array, axis=-1)/2
```

```python
img_gray_min_max.shape
```

```python
plt.imshow(img_gray_min_max, cmap='gray')
```

### Solution 3: Luma

```python
img_gray_luma = np.dot(img_array, [0.21, 0.72, 0.07])
img_gray_luma.shape
```

```python
plt.imshow(img_gray_luma, cmap='gray')
```

### Solution 4: Min

```python
img_gray_min = np.min(img_array, axis=-1)
```

### Solution 5: Max

```python
img_gray_max = np.max(img_array, axis=-1)
```

### Comparison

```python
plt.figure()

#subplot(r,c) provide the no. of rows and columns
f, axarr = plt.subplots(1, 5) 
axarr[0].imshow(img_gray, cmap="gray", vmin=0, vmax=255)
axarr[1].imshow(img_gray_min_max, cmap="gray", vmin=0, vmax=255)
axarr[2].imshow(img_gray_luma, cmap="gray", vmin=0, vmax=255)
axarr[3].imshow(img_gray_min, cmap="gray", vmin=0, vmax=255)
axarr[4].imshow(img_gray_max, cmap="gray", vmin=0, vmax=255)

# hide axis tickers
for i in range(5):
    # Turn off tick labels
    axarr[i].set_yticklabels([])
    axarr[i].set_xticklabels([])

plt.show()
```
