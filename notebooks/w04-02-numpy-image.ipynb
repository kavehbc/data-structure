```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = Image.open("data/Lenna.png").resize((250, 250))
img
```

```python
img.size
```

```python
img_array = np.array(img)
img_array
```

```python
type(img_array)
```

```python
img_array.shape
```

```python
plt.imshow(img_array)
```

```python
plt.imshow(img_array[:,:,0], cmap="Reds", vmin=0, vmax=255)
```

```python
plt.imshow(img_array[:,:,1], cmap="Greens", vmin=0, vmax=255)
```

```python
plt.imshow(img_array[:,:,2], cmap='Blues', vmin=0, vmax=255)
```

```python
plt.imshow(img_array[100:200,100:200,:])
```

```python
img_bw = np.average(img_array, axis=-1)
img_bw.shape
```

```python
plt.imshow(img_bw, cmap='gray')
```

```python
img_bw_min_max = np.max(img_array, axis=-1)/2 + np.min(img_array, axis=-1)/2
```

```python
img_bw_min_max.shape
```

```python
plt.imshow(img_bw_min_max, cmap='gray')
```

```python
img_bw_lum = np.dot(img_array, [0.21, 0.72, 0.07])
img_bw_lum.shape
```

```python
plt.imshow(img_bw_lum, cmap='gray')
```
